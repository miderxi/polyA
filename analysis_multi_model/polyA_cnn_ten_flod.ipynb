## 1. 第一个polya信号数据的5折交叉验证

```python
# 首先使用AATAGA单独效果
import ps
from sklearn.model_selection import train_test_split
import multiprocessing
import importlib
import os
import pandas as pd
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
```

```python
#
polys="AATAGA AATATA CATAAA GATAAA AGTAAA ACTAAA TATAAA AATACA AAAAAG AAGAAA ATTAAA AATAAA".split(" ")
poly = polys[0]
folds = 5

from keras import regularizers
from keras.models import Sequential
from keras.layers import Conv1D,MaxPool1D,Flatten,Dense,Dropout,Activation
from keras.optimizers        import Adam  
importlib.reload(ps)
def my_model():
    model = Sequential()
    model.add(Conv1D(64,7,padding='same',activation='relu',input_shape=(200,4),kernel_regularizer=regularizers.l1_l2(0.001,0.001)))
    model.add(MaxPool1D(2))
    model.add(Conv1D(64,7,padding='same',activation='relu'))
    model.add(MaxPool1D(2))
    
    model.add(Flatten())
    model.add(Dense(128,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(1,activation='sigmoid'))

    model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
    return model   

def my_average(df):
    df = pd.DataFrame(df[df[:,3]>0])
    return df.describe().iloc[1,:].to_numpy()

#生成数据
X,y = ps.get_onehot_data(poly)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)

#模型训练与评价
model = my_model()
history = model.fit(X_train,y_train,epochs=100,verbose=0,validation_split=0.25)
ps.showhist(history)
print( ps.assess(y_test,model(X_test)))

from sklearn.model_selection import StratifiedKFold
import numpy as np
folds=5
result = np.zeros((folds,5))
kfold = StratifiedKFold(n_splits=folds,random_state=23,shuffle=True)
for i,(train, test) in enumerate(kfold.split(X, y)):
    X_train,y_train,X_test,y_test = X[train],y[train],X[test],y[test]
    model = my_model()
    model.fit(X_train,y_train,epochs=40,verbose=0)
    temp = ps.assess(y_test,model(X_test))
    result[i] = np.array(list(temp.values()))
    print(f'{i}fold; {temp}')
print(my_average(result))

```

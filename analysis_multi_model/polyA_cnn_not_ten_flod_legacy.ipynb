```python
# 5-fold validation
# 0.8 for train,0.2 for test

import tensorflow as tf
from keras.callbacks import EarlyStopping
import numpy as np
from importlib import reload
import ps
reload(ps)
import multiprocessing
np.random.seed(22)

polys="AATAGA AATATA CATAAA GATAAA AGTAAA ACTAAA TATAAA AATACA AAAAAG AAGAAA ATTAAA AATAAA".split(" ")


def run_model(poly='AATAGA'):
    re=[]#store y_test and y_test_predict result
    X,y = ps.get_onehot_data(poly)
    from sklearn.model_selection import KFold
    kfold = KFold(n_splits=5, shuffle=True,random_state=22)
    index = kfold.split(X,y)
    
    for train_index,test_index in index:
        def work(conn,X,y,train_index,test_index):
            X_train,X_test,y_train,y_test = X[train_index],X[test_index],y[train_index],y[test_index]
            model = ps.cnn_model()
            model.fit(X_train,y_train,epochs=100,verbose=0)

            y_test_prob = model.predict(X_test)
            conn.send([y_test,y_test_prob])
            conn.close()
            
        parent_conn, child_conn = multiprocessing.Pipe()
        p = multiprocessing.Process(target=work,args=(child_conn,X,y,train_index,test_index))
        p.start()
        re.append(parent_conn.recv())
        p.join()
    return re

for index,poly in enumerate(polys):
    print(index,':',poly)
    result = run_model(poly)
    ps.assess(np.array([a for a,b in result]).reshape(-1),
              np.array([b for a,b in result]).reshape(-1),poly_name=poly)
    
 
```
